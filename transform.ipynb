```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "2019_WHR.csv"
world_happiness_df = pd.read_csv(csv_file)
world_happiness_df.head()
```

```python
csv_file2 = "world_1.csv"
worldPop_df = pd.read_csv(csv_file2,encoding='latin1')
worldPop_df.head()
```

```python
new_world_happiness_df= world_happiness_df[['Overall rank','Country or region','Score']].copy()
new_world_happiness_df.head(156)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(new_world_happiness_df)
```

```python
new_world_pop_df= worldPop_df[['Country Name','2019']].copy()
new_world_pop_df.head()
```

```python
# TRANSFORM
```

```python
new_world_happiness_df = new_world_happiness_df.rename(columns = {'Overall rank': 'overall_rank', 'Country or region': 'country_name', 'Score': 'score'})
new_world_happiness_df.head()
```

```python
new_world_pop_df = new_world_pop_df.rename(columns = {'Country Name': 'country_name', '2019': '2019_population'})
new_world_pop_df.head()
```

```python
new_world_happiness_df.to_csv('world_happiness.csv', index = False)
```

```python
new_world_pop_df.to_csv('world_population.csv', index = False)
```
