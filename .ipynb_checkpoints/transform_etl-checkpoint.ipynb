```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgress@localhost:5432/happiness')
connection = engine.connect()
```

```python
# dataframe
happiness_population = pd.read_sql('SELECT happiness.country_name, happiness.overall_rank, round(happiness.score,2) AS score, population."2019_population" from happiness left join population on population.country_name = happiness.country_name order by happiness.overall_rank', connection)
happiness_population.head()
```

```python
happiness_population2 = pd.read_sql('SELECT * FROM happiness_population', connection)
happiness_population2.head()
```

```python
 # LOAD
happiness_population2.to_sql("Imported_Joined_Table", connection)
```

```python
new_world_pop_df.to_sql("new_world_pop", connection)
```

```python
new_world_happiness_df.to_sql("new_world_happiness", connection)
```

```python
join_tables = pd.read_sql('select new_world_happiness.country_name, new_world_happiness.overall_rank, new_world_happiness.score as score, new_world_pop."2019_population" from new_world_happiness left join new_world_pop on new_world_pop.country_name = new_world_happiness.country_name order by new_world_happiness.overall_rank', connection)
join_tables
```

```python
join_tables.to_sql("Imported_Joined_Table_2", connection)
```
