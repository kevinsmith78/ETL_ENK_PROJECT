```python
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import gmaps.datasets

# Import API key
from api_keys import g_key
```

```python
csv_file = "happiness_overallRank.csv"
t_world_happiness_df = pd.read_csv(csv_file)
t_world_happiness_df.head()
```

```python
csv_file = "country_lat_long.csv"
world_lat_df = pd.read_csv(csv_file)
world_lat_df.head()
```

```python
new_world_df=world_lat_df[['country','latitude','longitude']]
new_world_df.head()
```

```python
t_new_world_df = new_world_df.rename(columns = {'country': 'country_name'})
t_new_world_df.head()
```

```python
merged_countries=pd.merge(t_world_happiness_df,t_new_world_df, on='country_name', how='left')
merged_countries
```

```python
t_new_world_heat_df= merged_countries[['country_name','overall_rank','score','latitude','longitude']].copy()
t_new_world_heat_df
```

```python
gmaps.configure(api_key=g_key)
```

```python
locations=t_new_world_heat_df[["latitude", "longitude"]]
locations
```

```python
locations_1= locations.fillna(0)
locations_1
```

```python
rank=t_new_world_heat_df[["score"]]
rank
```

```python
rank_1= rank.fillna(0.0)
rank_1.dtypes
```

```python
fig = gmaps.figure()
gheat_layer=gmaps.heatmap_layer(locations_1, weights=rank_1.score, dissipating=False, max_intensity=10, point_radius=5)
fig.add_layer(gheat_layer)
fig
```

```python
fig= gmaps.figure()
markers = gmaps.marker_layer(locations_1)
fig.add_layer(markers)
fig
```
